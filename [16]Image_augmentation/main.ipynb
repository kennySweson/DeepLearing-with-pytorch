```python
import torch
import torchvision
from torch import nn
from d2l import torch as d2l
from matplotlib import pyplot as plt
from PIL import Image
```

```python
img = Image.open('/Users/sunweixiang/Desktop/DeepLearing-with-pytorch/[16]Image_augmentation/cat.jpeg')
plt.imshow(img), img.size
```

```python
def apply(img, aug, num_rows=2, num_cols=4, scale=1.5):
    Y = [aug(img) for _ in range(num_cols * num_rows)]
    d2l.show_images(Y, num_rows, num_cols)
```

```python
apply(img, torchvision.transforms.RandomHorizontalFlip())
```

```python
apply(img, torchvision.transforms.RandomResizedCrop(
    (200,200), # 目标大小
    scale=(0.1, 1), # 面积范围
    ratio=(0.5, 2) # 区域高宽比
))
```

```python
apply(img, torchvision.transforms.ColorJitter(
    brightness=0.8, # 范围1+-0.5
    contrast=0, # 对比度
    saturation=0, # 饱和度
    hue=0 # 色调
))
```

```python
apply(img, torchvision.transforms.ColorJitter(
    brightness=0, # 亮度
    contrast=0, # 对比度
    saturation=0, # 饱和度
    hue=0.5 # 色调
))
```

不同增广方法的结合

```python
apply(img, torchvision.transforms.Compose([
    torchvision.transforms.RandomHorizontalFlip(),
    torchvision.transforms.ColorJitter(
        brightness=0.5,
        contrast=0.5,
        saturation=0.5,
        hue=0.5
    )
]))
```
